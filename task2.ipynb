```python
import pandas as pd
#importing data set
csv_file = '/content/drive/My Drive/DataScience/pima-indians-diabetes.csv'
pima_data = pd.read_csv(csv_file)
```

```python
new_column_names = [
    'NumberOfTimesPregnant',
    'PlasmaGlucose2H',
    'DiastolicBloodPressure',
    'TricepsSkinfoldThickness',
    '2HourSerumInsulin',
    'BMI',
    'DiabetesPedigreeFunction',
    'Age',
    'ClassVariable'
]
pima_data.columns = new_column_names
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Check for missing values
missing_values = pima_data.isnull().sum()
print("Missing Values:")
print(missing_values)

# Summary statistics
summary_stats = pima_data.describe()
print("\nSummary Statistics:")
print(summary_stats)

# Explore the distribution of the target variable 'ClassVariable'
sns.countplot(x='ClassVariable', data=pima_data, palette='viridis')
plt.xlabel('ClassVariable')
plt.ylabel('Count')
plt.title('Distribution of ClassVariable in Pima Indians Diabetes Dataset')
plt.show()

# Explore relationships between variables using pairplot
sns.pairplot(pima_data, hue='ClassVariable', palette='husl')
plt.suptitle('Pairplot of Variables in Pima Indians Diabetes Dataset', y=1.02)
plt.show()
```
